```python
# Clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data_iris = iris.data
```

```python
# K-Mean Clustering
from sklearn.cluster import KMeans
```

```python
kmns = KMeans(n_clusters = 3)
```

```python
kmns.fit(data_iris)
```

```python
labels = kmns.predict(data_iris)
```

```python
ctn = kmns.cluster_centers_
```

```python
plt.scatter(data_iris[:,2], data_iris[:,3], c = labels)
plt.scatter(ctn[:,2], ctn[:,3], marker='o', color='red', s=120)
plt.xlabel('Petal Length (cms)')
plt.ylabel('Petal Width (cms)')
plt.show()
```

```python
kmns.inertia_
```

```python
# Calculate inertia for different values of K
inertia_list =[]
for i in range(1, 11):
    kmns = KMeans(n_clusters=i, random_state=44)
    kmns.fit(data_iris)
    inertia_list.append(kmns.inertia_)
```

```python
plt.plot(range(1, 11), inertia_list, color='green', marker='o')
plt.xlabel('Number of K')
plt.ylabel('Inertia')
plt.show()
```

```python
# DBSCAN
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps=0.7, min_samples=4)
```

```python
dbs.fit(data_iris)
```

```python
labels = dbs.labels_
```

```python
plt.scatter(data_iris[:,2], data_iris[:,3], c = labels)
plt.show()
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
hr = linkage(data_iris, method='complete')
```

```python
dnd = dendrogram(hr)
```

```python
labels = fcluster(hr, 4, criterion='distance')
```

```python
plt.scatter(data_iris[:,2], data_iris[:,3], c=labels)
plt.show()
```
